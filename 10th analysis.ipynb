```python
# Import libraries
import pandas as pd
import pyreadstat as pyr

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from wrangle2021 import Wrangler
```

```python
# Instantiate wrangle class
wrangle = Wrangler()
```

```python
# load data and rename columns
data = wrangle.wrangle(columns = ['REGION','Q16A', 'Q16B'], rename = ['State', 'Was media coverage fair?', 'Offered bribe for vote?'])
```

```python
# Remove noisy values
df_new = wrangle.remove(columns = ['State', 'Was media coverage fair?', 'Offered bribe for vote?'], df_new=data)
```

```python
# Get frequency of values
wrangle.get_frequency(columns = ['Was media coverage fair?', 'Offered bribe for vote?'], df_new=df_new)
```

```python
fig = plt.figure()

# Needed to add spacing between 1st and 2nd row
# Add a margin between the main title and sub-plots
fig.subplots_adjust(hspace=0.4, top=2)

# Add the main title
fig.suptitle("Media and Bribery during the last Nigerian General Election", y=2.2, fontsize=15)


# Add the subplots
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)




# Plot for axis 1-4
(df_new['Was media coverage fair?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax1, color='green')
(df_new['Offered bribe for vote?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax2, color='green')


# Add the text for each subplot
ax1.title.set_text('Was media coverage fair?')
ax2.title.set_text('Offered bribe for vote?')


# label
plt.xlabel('Percentage [%]')

plt.show()
```
