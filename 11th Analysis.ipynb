```python
# Import libraries
import pandas as pd
import pyreadstat as pyr

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from wrangle2021 import Wrangler
```

```python
# Instantiate wrangle class
wrangle = Wrangler()
```

```python
# load data and rename columns
data = wrangle.wrangle(columns = ['REGION','Q17A', 'Q17B', 'Q17C', 'Q17D'], rename = ['State', 'Inaccurately counted votes?', 'People voted more than once?', 'Can the powerful find out your vote?', 'Did you fear political intimidation or violence?'])
```

```python
# Get frequency of values
wrangle.get_frequency(columns = ['Inaccurately counted votes?', 'People voted more than once?', 'Can the powerful find out your vote?', 'Did you fear political intimidation or violence?'], df_new=data)
```

```python
fig = plt.figure()

# Needed to add spacing between 1st and 2nd row
# Add a margin between the main title and sub-plots
fig.subplots_adjust(hspace=0.4, top=2)

# Add the main title
fig.suptitle("2019 General Election", y=2.2, fontsize=15)


# Add the subplots
ax1 = fig.add_subplot(4,1,1)
ax2 = fig.add_subplot(4,1,2)
ax3 = fig.add_subplot(4,1,3)
ax4 = fig.add_subplot(4,1,4)




# Plot for axis 1-4
(data['Inaccurately counted votes?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax1, color='green')
(data['People voted more than once?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax2, color='green')
(data['Can the powerful find out your vote?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax3, color='green')
(data['Did you fear political intimidation or violence?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax4, color='green')

# Add the text for each subplot
ax1.title.set_text('Were votes inaccurately counted?')
ax2.title.set_text('Did people vote more than once?')
ax3.title.set_text('Can the powerful find out your vote?')
ax4.title.set_text('Did you fear political intimidation or violence?')

# label
plt.xlabel('Percentage [%]')

plt.show()
```
