```python
# Import libraries
import pandas as pd
import pyreadstat as pyr

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from wrangle2021 import Wrangler
```

```python
# Instantiate wrangle class
wrangle = Wrangler()
```

```python
# Rename columns
data = wrangle.wrangle(columns=['REGION','Q21'],rename=['State', 'Support Democracy'])
```

```python
# Get frequency
wrangle.get_frequency(columns=['Support Democracy'], df_new=data)
```

```python
# Remove values not needed
df_new = wrangle.remove(columns=['Support Democracy'], df_new=data)
```

```python
# Remove part of the data not needed
df_new['Support Democracy'] = df_new['Support Democracy'].str.split(':', expand=True)[1]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
(df_new['Support Democracy'].value_counts(normalize=True).sort_values()*100).plot(kind='pie', autopct='%1.0f%%', ax=ax, title='Do you support democracy?')
ax.set(ylabel=None)
plt.show();
```
