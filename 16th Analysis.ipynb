```python
# Import libraries
import pandas as pd
import pyreadstat as pyr

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from wrangle2021 import Wrangler
```

```python
# Instantiate wrangle class
wrangle = Wrangler()
```

```python
# Get data and rename columns
data = wrangle.wrangle(columns=['REGION','Q35A', 'Q35B', 'Q35C', 'Q35D'],rename=['State', 'Minimum age for presidential candidate', 'Suggested minimum age for presidential candidate', 'Maximum age for presidential canditate', 'Suggested maximum age for presidential candidate'])
```

```python
# Get frequency
wrangle.get_frequency(columns=['Minimum age for presidential candidate', 'Suggested minimum age for presidential candidate', 'Maximum age for presidential canditate', 'Suggested maximum age for presidential candidate'], df_new=data)
```

```python
# Remove values not needed
df_new = wrangle.remove(columns=['Minimum age for presidential candidate', 'Suggested minimum age for presidential candidate', 'Maximum age for presidential canditate', 'Suggested maximum age for presidential candidate'], df_new=data)
```

```python
fig = plt.figure(figsize=(15, 10))

# Needed to add spacing between 1st and 2nd row
# Add a margin between the main title and sub-plots
fig.subplots_adjust(hspace=0.2, wspace=0.3, top=2)

# Add the main title
fig.suptitle("Minimum and maximum ages for presidential candidates", y=2.2, fontsize=15)


# Add the subplots
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)




# Plot for axis 1-3
(df_new['Minimum age for presidential candidate'].value_counts(normalize=True).
 sort_values()*100).plot(kind='pie', autopct='%1.0f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                         ax=ax1, colors = ['#B7C3F3','#8EB897'], title='Should there be a minimum age for a presidential candidate?')
ax1.set(ylabel=None)
(df_new['Suggested minimum age for presidential candidate'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax2, color='green')
(df_new['Maximum age for presidential canditate'].value_counts(normalize=True).
 sort_values()*100).plot(kind='pie', autopct='%1.0f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, 
                         ax=ax3, colors = ['#B7C3F3','#8EB897'], title='Should there be a maximum age for a presidential candidate?')
ax3.set(ylabel=None)
(df_new['Suggested maximum age for presidential candidate'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax4, color='green')

# Add the text for each subplot
ax1.title.set_text('Should there be a minimum age for a presidential candidate?')
ax2.title.set_text('Suggested minimum age for presidential candidate')
ax3.title.set_text('Should there be a maximum age for a presidential candidate?')
ax4.title.set_text('Suggested maximum age for presidential candidate')


# label
plt.xlabel('Percentage [%]')

plt.show()
```
