```python
# Import libraries
import pandas as pd
import pyreadstat as pyr

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from wrangle2021 import Wrangler
```

```python
# Instantiate wrangle class
wrangle = Wrangler()
```

```python
# wrangle data
data = wrangle.wrangle(columns=['REGION','Q18A', 'Q18B', 'Q18C'], rename=['State', 'Police assist people to cast ballot?', 'Voter intimidation?', 'Accurate results?'])
```

```python
# Get frequency of values
wrangle.get_frequency(columns = ['Police assist people to cast ballot?', 'Voter intimidation?', 'Accurate results?'], df_new=data)
```

```python
df_new = wrangle.remove(columns=['Police assist people to cast ballot?', 'Voter intimidation?', 'Accurate results?'], df_new=data)

wrangle.get_frequency(columns = ['Police assist people to cast ballot?', 'Voter intimidation?', 'Accurate results?'], df_new=df_new) 
```

```python
fig = plt.figure()

# Needed to add spacing between 1st and 2nd row
# Add a margin between the main title and sub-plots
fig.subplots_adjust(hspace=0.4, top=2)

# Add the main title
fig.suptitle("2019 General Election", y=2.2, fontsize=15)


# Add the subplots
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)




# Plot for axis 1-3
(df_new['Police assist people to cast ballot?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax1, color='green')
(df_new['Voter intimidation?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax2, color='green')
(df_new['Accurate results?'].value_counts(normalize=True).sort_values()*100).plot(kind='barh', ax=ax3, color='green')

# Add the text for each subplot
ax1.title.set_text('Did the police assist people to cast their ballot?')
ax2.title.set_text('Was there voter intimidation?')
ax3.title.set_text('Were results accurate?')


# label
plt.xlabel('Percentage [%]')

plt.show()
```
